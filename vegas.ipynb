```python
# coding: utf-8
import json

import requests
import time
import pandas as pd
import matplotlib.pyplot as plt
import os

# 绘制滤波前后的时间序列图
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为宋体或其他已安装的字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题

PATH = 'D:/All About Learning/Quant/mexc'
```

```python
with open(f'{PATH}pairs.json', 'r') as f:
    pairs = json.load(f)
```

```python
btc = pd.read_csv(f'{PATH}/BTCUSDT.csv', index_col='open_time')
btc = btc.sort_index()
btc
```

```python
for p in pairs:
    pair = pd.read_csv(f'{PATH}/{p}.csv', index_col='open_time')
    pair = pair.sort_index()

    pair['btc'] = btc['open']

    pair['/btc'] = pair['open']/pair['btc']
    pair['MA12'] = pair['/btc'].ewm(span=12, adjust=False).mean()
    pair['MA144'] = pair['/btc'].ewm(span=144, adjust=False).mean()
    pair['MA169'] = pair['/btc'].ewm(span=169, adjust=False).mean()
    pair['MA576'] = pair['/btc'].ewm(span=576, adjust=False).mean()
    pair['MA676'] = pair['/btc'].ewm(span=676, adjust=False).mean()

    pair.to_csv(f'{PATH}/vBTC/{p}.csv')

    print(f'Finish {p}')
```
